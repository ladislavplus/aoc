```python
import utils

data = utils.get_aoc_input(2021, 9).splitlines()
print(len(data), "lines")
```

```python
# aoc 2021 day 9 part 1
def part1(data):
    heightmap = [[int(c) for c in line] for line in data]
    rows, cols = len(heightmap), len(heightmap[0])
    
    def is_low_point(r, c):
        current = heightmap[r][c]
        for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols:
                if heightmap[nr][nc] <= current:
                    return False
        return True
    
    risk_level_sum = 0
    for r in range(rows):
        for c in range(cols):
            if is_low_point(r, c):
                risk_level_sum += heightmap[r][c] + 1
                
    return risk_level_sum
print("Part 1:", part1(data))
```

```python
# part 2
def part2(data):
    heightmap = [[int(c) for c in line] for line in data]
    rows, cols = len(heightmap), len(heightmap[0])
    visited = [[False] * cols for _ in range(rows)]
    
    def dfs(r, c):
        if r < 0 or r >= rows or c < 0 or c >= cols:
            return 0
        if visited[r][c] or heightmap[r][c] == 9:
            return 0
        visited[r][c] = True
        size = 1
        for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            size += dfs(r + dr, c + dc)
        return size
    
    basin_sizes = []
    for r in range(rows):
        for c in range(cols):
            if not visited[r][c] and heightmap[r][c] != 9:
                basin_size = dfs(r, c)
                basin_sizes.append(basin_size)
    
    basin_sizes.sort(reverse=True)
    return basin_sizes[0] * basin_sizes[1] * basin_sizes[2]

print("Part 2:", part2(data))
```
